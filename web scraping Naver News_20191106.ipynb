```python
title_list = []
body_list = []
for i in range(len(news_link)):
    req1 = urllib.request.urlopen(news_link[i])
    data1 = req1.read()
    soup1 = BeautifulSoup(data1, "html.parser")
    title = soup1.find("h3",{"id": "articleTitle"}).get_text()
    body = soup1.find("div", {"id" : "articleBodyContents"}).get_text().replace("\n","").replace("// flash 오류를 우회하기 위한 함수 추가 function _flash_removeCallback() {}", "")

    title_list.append(title)
    body_list.append(body)

    
title_list
```
